```python
import pandas as pd
import re
use_cases = re.compile('use.cases|use.case', re.IGNORECASE)
object_re = re.compile("object.diagram", re.IGNORECASE)
```

```python
file_types = pd.read_csv("./db/Project_FileTypes_V2.0.csv")
uml_files = pd.read_csv("./db/UMLFiles_List_V2.0.csv")
```

```python
pd.set_option('display.max_colwidth', None)
use_cases_df = uml_files[uml_files['Model Link - Github'].str.contains(use_cases)==True]['Model Link - Github']
uml_files[uml_files['Model Link - Github'].str.contains(use_cases)==True]
```

```python
get_user = re.compile("^https:\/\/www\.github\.com\/([a-zA-Z0-9_-]+)\/")
users = []
BOLD = '\033[1m'
END = '\033[0m'
count = 0
print(BOLD+"Unique users:\n"+END)
for link in use_cases_df:
    s = re.search(get_user, link)
    user = s.group(1)
    if user in users:
        continue
    else:
        users.append(user)
        count+=1
        if count > 10 and count < 550:
            if count == 11:
                print("...\n")
        else:
            print(f'{count}.\t{link.replace(" ", "%20")}\n')
print(BOLD + f'\n{count} unique users' + END)
```
